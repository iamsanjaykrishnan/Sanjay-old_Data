```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

## Beta Distribution

```python
probablity=np.linspace(0,1,100)
a_0=1
b_0=1
pdf_0=beta.pdf(probablity, a_0, b_0)
plt.plot(probablity,pdf_0)
```

```python
mean, var = beta.stats(a_0, b_0, moments='mv')
print(mean)
print(var)
```

## Evidence
##### 1 out of 10 people have cancer

```python
a_1=a_0+1
b_2=b_0+10-1
pdf_1=beta.pdf(probablity, a_1, b_2)
plt.plot(probablity,pdf_1)
```

```python
mean, var, skew, kurt = beta.stats(a_1, b_2, moments='mvsk')
print(mean)
print(var)
```

```python
a_1=a_1+1.0
b_2=b_2+10-1.0
pdf_1=beta.pdf(probablity, a_1, b_2)
plt.plot(probablity,pdf_1)
mean, var, skew, kurt = beta.stats(a_1, b_2, moments='mvsk')
print(mean)
print(var)
```
