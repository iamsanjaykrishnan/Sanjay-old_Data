```python
%matplotlib inline 
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
tf.logging.set_verbosity(tf.logging.ERROR)
```

```python
omega=np.linspace(0,16,num=60)
omega=omega.reshape([-1,1])
omega_highres=np.linspace(0,10*3.14,num=500)
omega_highres=omega_highres.reshape([-1,1])
sin_sig_with_noise=+np.random.normal(np.sin(omega),0.2)
```

```python
plt.scatter(omega,sin_sig_with_noise)
plt.show()
```

```python
tf.reset_default_graph()
tf_input=tf.placeholder(shape=[None,1],dtype=tf.float32)
output_label=tf.placeholder(shape=[None,1],dtype=tf.float32)

#hiddenlayer=tf.contrib.bayesflow.layers.dense_flipout(units=200,inputs=tf_input,)
hiddenlayer=tf.layers.dense(units=20,inputs=tf_input)
hiddenlayer=tf.nn.tanh(hiddenlayer)
#hiddenlayer=tf.contrib.bayesflow.layers.dense_flipout(units=200,inputs=tf_input,)
hiddenlayer=tf.layers.dense(units=20,inputs=tf_input)
hiddenlayer=tf.nn.tanh(hiddenlayer)
#Output=tf.contrib.bayesflow.layers.dense_flipout(units=1,inputs=hiddenlayer)
Output=tf.layers.dense(units=1,inputs=hiddenlayer)

loss1=tf.losses.mean_squared_error(output_label,Output)#+ 0.001*tf.reduce_sum(tf.get_collection(tf.GraphKeys.REGULARIZATION_LOSSES))

Optimzer1=tf.train.AdamOptimizer().minimize(loss1)

loss2=tf.losses.mean_squared_error(output_label,Output)+0.001*sum(tf.get_collection(tf.GraphKeys.REGULARIZATION_LOSSES))
Optimzer2=tf.train.AdamOptimizer().minimize(loss2)

loss3=tf.losses.mean_squared_error(output_label,Output)
Optimzer3=tf.contrib.bayesflow.optimizers.SGLDOptimizer( learning_rate=0.0001,preconditioner_decay_rate=0.7,
                                                        num_pseudo_batches=1,burnin=20000).minimize(loss3)

sess=tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for _ in range(0,10000):
    sess.run(Optimzer1,feed_dict={tf_input:omega,output_label:sin_sig_with_noise})
   
    
value=sess.run(Output,feed_dict={tf_input:omega_highres})
plt.plot(omega_highres,value)

plt.scatter(omega,sin_sig_with_noise)
plt.xlim((0,10))
plt.ylim((-2,2))
```
